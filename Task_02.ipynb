```python
import pandas as pd
```

```python
path="/content/drive/MyDrive/DL Lab datasets/Au_nanoparticle_dataset.csv"
df=pd.read_csv(path)
```

```python
df.head()
```

```python
# New data frame
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
print(new_df.head())
```

```python
# Display first 20 samples
print(new_df.head(20))
```

```python
# Mean
mean_values = new_df.mean()

# Standard deviation
std_values = new_df.std()

# Quartiles
quartiles = new_df.quantile([0.25, 0.5, 0.75])

print("Mean values:\n", mean_values)
print("\nStandard deviation values:\n", std_values)
print("\nQuartiles:\n", quartiles)
```

```python
# Plot histogram
import matplotlib.pyplot as plt

features = ['N_total', 'N_bulk', 'N_surface', 'R_avg']

plt.figure(figsize=(20, 4))  # Wide figure to fit 4 plots in one row

for i, feature in enumerate(features):
    plt.subplot(1, 4, i+1)
    plt.hist(new_df[feature], bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Scatter plots and Histograms
import seaborn as sns

sns.pairplot(new_df)
plt.suptitle('Pairplot of Au Nano Particle Features', y=1.02)
plt.show()
```

```python
g = sns.PairGrid(new_df)

# Diagonal
g.map_diag(lambda x, **kwargs: sns.histplot(x, kde=True, **kwargs))

plt.suptitle('Diagonal: Histogram + KDE', y=1.02)
plt.show()
```

```python
g = sns.PairGrid(new_df)

# Lower triangle
g.map_lower(sns.kdeplot)

plt.suptitle('Lower triangle: Bivariate KDE plots', y=1.02)
plt.show()
```
